```python
def binomial_table(n):
    """
    Build a DP table containing binomial coefficients up to n.
    table[i][j] = C(i, j) for 0 <= j <= i <= n
    """
    # initialize table with zeros
    table = []

    # Create rows one by one
    for i in range(n + 1):           # i is the row index (0..n)
        row = [0] * (i + 1)          # a row of length i+1 filled with zeros
        table.append(row)
    
    for i in range(n + 1):
        table[i][0] = 1        # C(i,0) = 1
        table[i][i] = 1        # C(i,i) = 1
        for j in range(1, i):
            table[i][j] = table[i-1][j-1] + table[i-1][j]
    return table


# Example usage:
if __name__ == "__main__":
    n = 5
    table = binomial_table(n)
    # Print Pascal's triangle up to row n
    for i in range(n + 1):
        print(f"Row {i}: ", table[i])
    # Example: C(7, 3)
    print("C(5,3) =", table[5][3])
```
