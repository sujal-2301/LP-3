```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding = "ISO-8859-1")
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
```

```python
numeric_cols = ['QUANTITYORDERED', 'PRICEEACH', 'ORDERLINENUMBER', 'SALES', 'QTR_ID', 'MONTH_ID', 'YEAR_ID', 'MSRP']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df[numeric_cols] = scaler.fit_transform(df[numeric_cols])   # inplace replacement
X = df[numeric_cols].copy()

print(X)
```

```python
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(1, 10)

for k in K:
    kmeanModel = KMeans(n_clusters=k, random_state=42).fit(X)

    d = cdist(X, kmeanModel.cluster_centers_, metric='euclidean')   # (n_samples, k)
    nearest = np.min(d, axis=1)                                     # (n_samples,)
    avg_sq_dist = np.mean(nearest**2)
    distortions.append(avg_sq_dist)

    # distortions.append(kmeanModel.inertia_ / X.shape[0])

    inertias.append(kmeanModel.inertia_)

    mapping1[k] = distortions[-1]
    mapping2[k] = inertias[-1]
```

```python
print("Distortion values:")

for key, val in mapping1.items():

    print(f'{key} : {val}')

plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```

```python
print("Inertia values:")
for key, val in mapping2.items():
    print(f'{key} : {val}')

plt.plot(K, inertias, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```
