```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def f(x):
    return (x+3)**2
```

```python
def df(x):
    return 2 * (x + 3)
    
```

```python
def gradient(xi, lr, iters):
    x = xi
    x_hist = [x]

    for i in range(iters):
        x = x - lr * df(x)
        x_hist.append(x)
    return x, x_hist
    
```

```python
xi = 2
lr = 0.1
iters = 1000

x, x_hist = gradient(xi, lr, iters)

print("Minima found at x = {:.2f}".format(x))

```

```python
boundary = np.linspace(-5, 5, 100)
plt.plot(boundary, f(np.array(boundary)))
plt.plot(x_hist, f(np.array(x_hist)), 'rx-')
```
