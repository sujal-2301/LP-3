```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x+3)**2
```

```python
def df(x):
    return 2*(x+3)
```

```python
def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    for i in range(iterations):
        x = x - learning_rate * df(x)  # update step
        print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
    return x

starting_point = 2
learning_rate = 0.01
iterations = 1000

minimum = gradient_descent(starting_point, learning_rate, iterations)
print(f"\nLocal minimum occurs at x = {minimum:.4f}, f(x) = {f(minimum):.4f}")
```

```python
x_vals = np.linspace(-10, 2, 100)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, label="f(x) = (x + 3) ^ 2")
plt.scatter(minimum, f(minimum), color='red', label="Local Minimum")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent Visualization")
plt.legend()
plt.show()
```
